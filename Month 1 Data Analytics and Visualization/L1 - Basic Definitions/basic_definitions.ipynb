# Basic definitions

## Sample Space

**Sample space** is the collection of all outcomes of an experiment

### Single Coin toss

When we toss a coin once, there are two outcomes <br>
1) Heads<br>
2) Tails


Let $S$ denote the sample space, $H$ denote the outcome called Heads, and $T$ denote the outcome called tails.<br>
We write the sample space as<br>
$S = \{H, T\}$

### Two coin tosses

We are tossing one after the other. There are four possible outcomes. <br>
1) The first toss is heads and second is also heads. {HH} <br>
2) The first toss is heads and second is tails. {HT} <br>
3) The first toss is tails and second is heads. {TH} <br>
4) The first toss is tails and second is also tails. {TT} <br><br>

<div>

</div>

$S = \{HH, HT, TH, TT\}$

## Event

Any subset of the samples space is called an **event**

### Dice

The sample space is given by $S = \{1, 2, 3, 4, 5, 6\}.$<br>
Consider the following subsets of $S$.<br>
1) $A = \{1, 3, 5\}$ <br>
2) $B = \{1\}$<br>
3) $C = \{2, 4, 6\}$<br>
4) $D = \{1, 3, 5, 7\}$<br>

Of these, $A$, $B$, and $C$ are events. 

$D$ is not an event, since it is not a subset of $S$ - this is because $7$ is an element of $D$ but not that of $S$.

## Set operations

### Dice

Consider again the sample space of dice: $S = \{1, 2, 3, 4, 5, 6\}.$

Suppose person $A$ bets on $\{1, 3, 5\}$ and person $B$ bets of $\{1, 5, 6\}$. 

Question: What are the outcomes where both $A$ <font color='red'>**and**</font> $B$ will win?<br>
Answer: $\{1, 5\}$

Mathematically, we write <br>
$A \cap B = \{1, 5\}$

<div>

</div>

Question: What are the outcomes where $A$ <font color='red'>**or**</font> $B$ will win?<br>
Answer: $\{1, 3, 5, 6\}$

Mathematically, we write <br>
$A \cup B = \{1, 3, 5, 6\}$

## Probability

### Coin

The sample space for a coin toss is $\{H, T\}.$<br>
There are two outcomes that are equally likely. <br>
We expect half the tosses to be heads and half the tosses to be tails. <br>
Hence, the probability of heads = 1/2 and probability of tails = 1/2

$P[H] = 1/2$<br>
$P[T] = 1/2$

When we say $P[H]=1/2$, it means that if we keep on tossing a coin, the proportion of times it lands heads will approximately be 1/2.

### Dice

Sample space of dice: $S = \{1, 2, 3, 4, 5, 6\}.$

Let $A = \{2, 4, 6\}$ <br>
Since there are 3 outcomes out of 6, we have $P(A) = 3/6$

Let $B = \{1, 2\}$ <br>
Since there are 2 outcomes out of 6, we have $P(B) = 2/6$

What is the probability of both $A$ and $B$ happening?<br>
$P(A \cap B) = P(\{2\}) = \displaystyle\frac{1}{6}$

What is the probability of both $A$ or $B$ happening?<br>
$P(A \cup B) = P(\{1, 2, 4, 6\}) = \displaystyle\frac{4}{6}$

>Notice that we *cannot* simply add $P(A)$ and $P(B)$ to get $P(A \cup B)$<br>
Why is this? <br>
It is because $P(A \cap B) $ is counted twice, once in $P(A)$ and once again in $P(B)$. <br>
So the formula will be <br>
$P(A \cap B) = P(A) + P(B)$

What is the probability of the complement of $B$?<br>
$P(B^c) = P(\{3, 4, 5, 6\}) = \displaystyle\frac{4}{6}$ 

> Notice here that $P(B) + P(B^c) = 1$
